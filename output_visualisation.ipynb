```python
import matplotlib.pyplot as plt

# Example data structure
# data[graph_size][thread_count] = time
data = {
    6: {1: 0.000036413, 2:0.000190525, 4: 0.000135649, 8: 0.000254017, 16: 0.00182778},
    10: {1:0.000043103 , 2:0.000214905, 4: 0.000285256, 8:0.000250731, 16: 0.00262232},
    20: {1:0.000135429, 2: 0.000230863, 4: 0.000133124, 8:0.00012633 ,16:0.00204067},
    40: {1:0.000131899,2:0.000164477,4:0.000137517,8:0.00022089,16:0.00606961},
    80: {1:0.000484954,2:0.0004662,4:0.000356922,8:0.00047076,16:0.00830843}
}

for N, timings in data.items():
    plt.figure()
    thread_counts = list(timings.keys())
    times = list(timings.values())
    speed_up = [times[0] / time for time in times]
    parallel_efficiency = [100 * su / tc for su, tc in zip(speed_up, thread_counts)]

    # Plot Speed Up
    plt.subplot(1, 2, 1)
    plt.plot(thread_counts, speed_up, marker='o')
    plt.title(f'Speed Up for N={N}')
    plt.xlabel('Number of Threads')
    plt.ylabel('Speed Up')

    # Plot Parallel Efficiency
    plt.subplot(1, 2, 2)
    plt.plot(thread_counts, parallel_efficiency, marker='o')
    plt.title(f'Parallel Efficiency for N={N}')
    plt.xlabel('Number of Threads')
    plt.ylabel('Efficiency (%)')

    plt.tight_layout()
    plt.show()
```
